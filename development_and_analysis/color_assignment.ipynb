# Color Assignment


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans 
```

```python
image_path = "../output_videos/cropped_img.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show()
```

# Take top half of image (T-Shirt Color)

```python
top_half_image = image[0: int(image.shape[0]/2), :]
plt.imshow(top_half_image)
plt.show()
```

# Cluster the image into 2 clusters

```python
# Reshape the img into a 2d array 
image_2d = top_half_image.reshape(-1, 3)

#k-means clustering
kmeans = KMeans(n_clusters=2,random_state=0)
kmeans.fit(image_2d)

labels = kmeans.labels_

#reshape labels back to the og image shape
clustered_image = labels.reshape(top_half_image.shape[0], top_half_image.shape[1])
plt.imshow(clustered_image)
plt.show()
```

```python
corner_clusters = [clustered_image[0,0], clustered_image[0,-1], clustered_image[-1,0], clustered_image[-1,-1]]
non_player_cluster = max(set(corner_clusters), key=corner_clusters.count)
print(non_player_cluster)
```

```python
player_cluster = 1 - non_player_cluster
print(player_cluster)
```

```python
kmeans.cluster_centers_[player_cluster] #RGB
```
